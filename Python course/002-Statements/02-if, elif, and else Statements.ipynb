* **<code>if</code>** Statements in Python allows us to **tell the computer** to perform **alternative actions** based on a certain set of results, Something like:

"<font color='green'>Hey if this case happens</font>,  <font color='red'>perform some action</font> "

* We can then expand the idea further with <code>elif</code> and <code>else</code> statements, which allow us to tell the computer:

"Hey if this case happens, perform some action. 

**Else (elif)**, if another case happens, perform some other action. 

**Else**, if *none* of the above cases happened, perform this action."

Syntax format for <code>if</code> statements:

#### if else:

    if case1:
        perform action1
   
    else: 
        perform action2

#### if-elif-else:

    if case1:
        perform action1
    elif case2:
        perform action2
    else: 
        perform action3

## First Example

Let's see a quick example of this:

```python
x=2

if x==2:  # True is always true! it is as an example here. Basically, you can use the condition here based on your goal.
    print('It was true!')
```

Let's add in some else logic:

```python
x = 20 # False is always false!

if x==2:
    print('x was True!')
else:
    print('It will be printed in any case where x is not true')
```

```python
# example if x==10  

x=20

if x==2:
    print("x is true")
elif x==10:
    print("x is at least a number")  
else:
    print("In any case x is not true")
        
```

### Multiple Branches

Take note of how the <code>if</code>, <code>elif</code>, and <code>else</code> line up in the code. This can help you see what <code>if</code> is related to what <code>elif</code> or <code>else</code> statements.

```python
 loc = 'Bank'

if loc == 'Auto Shop':
    print('Welcome to the Auto Shop!')
elif loc == 'Bank':
    print('Welcome to the bank!')
else:
    print('Where are you?')
```

#### You  can put in as many <code>elif</code> statements as you want before you close off with an <code>else</code>.
