These strange terms show up as parameters in function definitions.

```python
def myfunc(a,b):
    return sum((a,b))*.05

myfunc(40,60)
```

This function returns 5% of the sum of **a** and **b**. In this example, **a** and **b** are *positional* arguments; that is, 40 is assigned to **a** because it is the first argument, and 60 to **b**. Since **b** is the second argument. Notice also that to work with multiple positional arguments in the `sum()` function we had to pass them in as a tuple.

## `**kwargs`

##### Before running the following function, can you undrestand what`**kwargs`does? Try call this function, that is, enter an input and see the output.

```python
def demo(**kwargs):
    print(kwargs)
demo(name="Humpty", location="Wall") # add another key and value and see what would happen
```

##### What about the following function?

```python
def myfunc(**kwargs):
    if 'fruit' in kwargs:
        print(f"My favorite fruit is {kwargs['fruit']}")  # review String Formatting and f-strings if this syntax is unfamiliar
    else:
        print("I don't like fruit")
        
myfunc(fruit='pineapple')
```
